### `__str__()`ã€`__repr__()`

````python
name=''
class Animal(object):
    def __init__(self,name):
        self.name=name
    def run(self):
        print('åŠ¨ç‰©ä¼šè·‘ğŸ¯')
    def __str__(self):
        return '%sä¼šè·‘'%self.name
    __repr__=__str__#åœ¨äº¤äº’æ¨¡å¼ä¸‹ä¼šè°ƒç”¨repr

print(Animal('éª†é©¼ğŸ«'))
````

### `__iter__()`
&emsp;&emsp;å¦‚æœæƒ³å°†ä¸€ä¸ªç±»ç”¨äºfor...inå¾ªç¯ï¼Œç±»ä¼¼listå’Œtupleä¸€æ ·ï¼Œå°±å¿…é¡»å®ç°ä¸€ä¸ª`__iter__()`æ–¹æ³•ã€‚è¯¥æ–¹æ³•è¿”å›ä¸€ä¸ªè¿­ä»£å¯¹è±¡ï¼ŒPythonçš„forå¾ªç¯ä¼šä¸æ–­çš„è°ƒç”¨è¯¥è¿­ä»£å¯¹è±¡çš„`__next__()`æ–¹æ³•ï¼Œè·å¾—å¾ªç¯çš„ä¸‹ä¸€ä¸ªå€¼ï¼Œç›´åˆ°é‡åˆ°StopIterationé”™è¯¯æ—¶é€€å‡ºå¾ªç¯ã€‚

````python
class Fib(object):
    def __init__(self):
        self.a,self.b=0,1
    def __iter__(self):
        return self
    def __next__(self):
        self.a,self.b=self.b,self.a+self.b
        if self.a>100000:
            raise StopIteration()
        return self.a

for i in Fib():
    print(i)
````

### `__getitem__()`
&emsp;&emsp;ä¸ŠèŠ‚ä¾‹å­ä¸­è™½ç„¶èƒ½å®ç°forå¾ªç¯ï¼Œå’Œlistæœ‰ç‚¹åƒï¼Œä½†ä¸èƒ½å°†å…¶ä½œä¸ºlistä½¿ç”¨ï¼Œæ— æ³•è¿›è¡Œç´¢å¼•ç­‰æ“ä½œã€‚è¦åƒlistä¸€æ ·æŒ‰ç…§ä¸‹æ ‡å–å‡ºå…ƒç´ ï¼Œéœ€è¦`__getitem__()`æ–¹æ³•ï¼Œä¾‹å¦‚ï¼š

````python
class Fib(object):
    def __getitem__(self, item):
        a,b=1,1
        for x in range(item):
            a,b=b,a+b
        return a
fib=Fib()
print(fib[3])
````

### `__getattr__()`
&emsp;&emsp;æ­£å¸¸æƒ…å†µä¸‹ï¼Œè°ƒç”¨ç±»çš„æ–¹æ³•æˆ–å±æ€§æ—¶ï¼Œå¦‚æœç±»çš„æ–¹æ³•æˆ–å±æ€§ä¸å­˜åœ¨å°±ä¼šæŠ¥é”™ï¼Œï¼Œä¸ºäº†é¿å…è¿™ç§é”™è¯¯ï¼ŒPythonæä¾›äº†ä¸€ä¸ª`__getattr__()`æ–¹æ³•ï¼ŒåŠ¨æ€è¿”å›ä¸€ä¸ªå±æ€§ï¼š

````python
class Student(object):
    def __init__(self,name):
        self.__name=name
    def __str__(self):
        return 'å­¦ç”Ÿå§“åæ˜¯%s'%self.__name
    def __getattr__(self, item):
        if item!='__name':
            return 'ä½ æŸ¥æ‰¾çš„å±æ€§ä¸å­˜åœ¨'

a=Student('å°æ™º')
print(a)
print(a.__name)#ç§æœ‰å±æ€§ï¼Œè¿”å›None
print(a.score)
````

### `__call__()`
&emsp;&emsp;ä¸€ä¸ªå¯¹è±¡å¯ä»¥æœ‰è‡ªå·±çš„å±æ€§å’Œæ–¹æ³•ï¼Œé™¤äº†è°ƒç”¨æ–¹æ³•å¤–ï¼Œè¿˜å¯ä»¥è°ƒç”¨å¯¹è±¡æœ¬èº«ï¼Œåªéœ€è¦å®šä¹‰ä¸€ä¸ª`__call__()`æ–¹æ³•ã€‚

````python
class Student(object):
    def __init__(self,name):
        self.__name=name
    def __call__(self, *args, **kwargs):
        return 'å­¦ç”Ÿå§“åæ˜¯%sï¼Œåˆ†æ•°æ˜¯%d'%(self.__name,args[0])
a=Student('å°æ™º')
print(a)
print(a(100))
````
